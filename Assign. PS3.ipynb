```python
import pandas as pd
import numpy as np

csv_path = "C:\\Users\\yogit\\Downloads\\wind.txt"
data = pd.read_csv(csv_path, sep='\s+', parse_dates=[[0, 1, 2]])

data = data.rename(columns={'Yr_Mo_Dy': 'Date'})
data.set_index('Date', inplace=True)

def fix_year(year):
    if year > 2000:
        year -= 100
    return year

data.index = data.index.map(lambda x: pd.Timestamp(fix_year(x.year), x.month, x.day))


data.index = pd.to_datetime(data.index)

missing_values_per_location = data.isnull().sum()

total_non_missing_values = data.count().sum()

mean_windspeed = data.mean().mean()

loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})


average_windspeed_january = data[data.index.month == 1].mean()

yearly_data = data.resample('Y').mean()

monthly_data = data.resample('M').mean()

weekly_data = data.resample('W').mean()

weekly_stats = weekly_data[:52].describe()

print("Missing values per location:\n", missing_values_per_location)
print("\nTotal non-missing values:", total_non_missing_values)
print("\nMean windspeed over all locations and times:", mean_windspeed)
print("\nLocation statistics:\n", loc_stats)
print("\nDay statistics:\n", day_stats)
print("\nAverage windspeed in January for each location:\n", average_windspeed_january)
print("\nYearly data:\n", yearly_data)
print("\nMonthly data:\n", monthly_data)
print("\nWeekly data:\n", weekly_data)
print("\nWeekly statistics for the first 52 weeks:\n", weekly_stats)
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

first_10_entries = chipo.head(10)


num_observations = chipo.shape[0]

num_columns = chipo.shape[1]

column_names = chipo.columns.tolist()

index_type = chipo.index

most_ordered_item = chipo['item_name'].value_counts().idxmax()

num_most_ordered_item = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()

most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()

total_items_ordered = chipo['quantity'].sum()

chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

item_price_type = chipo['item_price'].dtype

revenue = (chipo['quantity'] * chipo['item_price']).sum()

num_orders = chipo['order_id'].nunique()

average_revenue_per_order = revenue / num_orders

num_different_items = chipo['item_name'].nunique()

print("First 10 entries:\n", first_10_entries)
print("Number of observations:", num_observations)
print("Number of columns:", num_columns)
print("Column names:", column_names)
print("Index type:", index_type)
print("Most-ordered item:", most_ordered_item)
print("Number of most-ordered item:", num_most_ordered_item)
print("Most-ordered choice description:", most_ordered_choice_description)
print("Total items ordered:", total_items_ordered)
print("Item price type:", item_price_type)
print("Revenue:", revenue)
print("Number of orders:", num_orders)
print("Average revenue per order:", average_revenue_per_order)
print("Number of different items sold:", num_different_items)
```

Question 6

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "C:\\Users\\yogit\\Downloads\\us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(csv_path)

years = data['Year']
marriages_per_1000 = data['Marriages']  
divorces_per_1000 = data['Divorces']    

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_1000, label='Marriages_per_1000')
plt.plot(years, divorces_per_1000, label='Divorces_per_1000')
plt.xlabel('Year')
plt.ylabel('Per 1000 Rate')
plt.title('Number of Marriages and Divorces_per_1000 in the U.S. (1867-2014)')
plt.legend()
plt.grid(True)
plt.show()
```
